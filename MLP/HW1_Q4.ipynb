##4_2.Multilayer perceptron

###part1to3

```python
import tensorflow as tf
Ytrain=train['price'].values
Xtrain=train.loc[:, train.columns !='price']
Ytest=test['price'].values
Xtest=test.loc[:, test.columns !='price']
Ytrain=tf.convert_to_tensor(Ytrain)
Xtrain=tf.convert_to_tensor(Xtrain)
Ytest=tf.convert_to_tensor(Ytest)
Xtest=tf.convert_to_tensor(Xtest)
Xtrain
```

```python
from tensorflow.keras.layers import Dense
def r_squared(y, y_pred):
  residual = tf.reduce_sum(tf.square(tf.subtract(y, y_pred)))
  total = tf.reduce_sum(tf.square(tf.subtract(y, tf.reduce_mean(y))))
  r2 = tf.subtract(1.0, tf.divide(residual, total+1e-6))
  return r2


onelayer=tf.keras.Sequential([
    Dense(64,input_shape=(73,),activation='relu'),
    Dense(1,activation='linear')
    ])

onelayer.compile(loss='mse',optimizer='sgd',metrics=r_squared)
onelayer.summary()
```

```python
histonelayer=onelayer.fit(Xtrain,Ytrain,validation_data=(Xtest,Ytest),epochs=100,batch_size=10)
```

```python
twolayer=tf.keras.Sequential([
    Dense(64,input_shape=(73,),activation='relu'),
    Dense(64,activation='relu'),
    Dense(1,activation='linear')
    ])

twolayer.compile(loss='mse',optimizer='sgd',metrics=r_squared)
twolayer.summary()
```

```python
histtwolayer=twolayer.fit(Xtrain,Ytrain,validation_data=(Xtest,Ytest),epochs=100,batch_size=10)
```

```python
threelayer=tf.keras.Sequential([
    Dense(64,input_shape=(73,),activation='relu'),
    Dense(64,activation='relu'),
    Dense(64,activation='relu'),
    Dense(1,activation='linear')
    ])

threelayer.compile(loss='mse',optimizer='sgd',metrics=r_squared)
threelayer.summary()
```

```python
histthreelayer=threelayer.fit(Xtrain,Ytrain,validation_data=(Xtest,Ytest),epochs=100,batch_size=10)
```

```python
plt.figure(figsize=(10,10))
plt.plot(histonelayer.history['loss'])
plt.plot(histtwolayer.history['loss'])
plt.plot(histthreelayer.history['loss'])
plt.title('Train Loss Function')
plt.xlabel('Epochs')
plt.legend(['loss_onelayer','loss_twolayer','loss_threelayer'])
plt.figure(figsize=(10,10))
plt.plot(histonelayer.history['val_loss'])
plt.plot(histtwolayer.history['val_loss'])
plt.plot(histthreelayer.history['val_loss'])
plt.legend(['val_loss_onelayer','val_loss_twolayer','val_loss_threelayer'])
plt.title('Test Loss Function')
plt.xlabel('Epochs')
```

```python
plt.figure(figsize=(10,10))
plt.plot(histonelayer.history['r_squared'])
plt.plot(histtwolayer.history['r_squared'])
plt.plot(histthreelayer.history['r_squared'])
plt.title('Train R2 Score')
plt.xlabel('Epochs')
plt.legend(['one layer','two layer','three layer'])
plt.figure(figsize=(10,10))
plt.plot(histonelayer.history['val_r_squared'])
plt.plot(histtwolayer.history['val_r_squared'])
plt.plot(histthreelayer.history['val_r_squared'])
plt.legend(['one layer','two layer','three layer'])
plt.title('Test R2 Score')
plt.xlabel('Epochs')
plt.ylim([-8000, 100])
```

R2score:

It tells how close are data to the fitted regression line.

Highest score can be 1.0 and it indicates that the predictors perfectly accounts for variation in the target.

Score 0.0 indicates that the predictors do not account for variation in the target.

It can also be negative if the model is worse.

<!-- After 100 epochs validation R2score:
 * for model with one hidden layer we have: 0.25
 * for model with two hidden layers we have: 0.53
 * for model with three hidden layer we have: 0.68

So based on R2score three hidden layer network was the best.  -->


```python
r2train1 = histonelayer.history['r_squared'][-10:]
r2train2  = histtwolayer.history['r_squared'][-10:]
r2train3  = histthreelayer.history['r_squared'][-10:]

r2test1 = histonelayer.history['val_r_squared'][-10:]
r2test2  = histtwolayer.history['val_r_squared'][-10:]
r2test3  = histthreelayer.history['val_r_squared'][-10:]
print(f'Train R2 scores:\n1 layer = {np.mean(r2train1)}\n2 layer = {np.mean(r2train2)}\n3 layer = {np.mean(r2train3)}')
print(f'\ntest R2 scores:\n1 layer = {np.mean(r2test1)}\n2 layer = {np.mean(r2test2)}\n3 layer = {np.mean(r2test3)}')
```

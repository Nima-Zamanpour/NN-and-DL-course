```python
import numpy as np
def calculate_end_position(row):
    start_pos = row['answer_start']
    text = row['text']
    if np.isnan(start_pos):
        return 0
    else:
        end_pos = start_pos + len(text)-1
        return end_pos
df_train['answer_end'] = df_train.apply(calculate_end_position, axis=1)
df_valid['answer_end'] = df_valid.apply(calculate_end_position, axis=1)
df_test['answer_end'] = df_test.apply(calculate_end_position, axis=1)
```

```python
from transformers import AutoConfig, AutoTokenizer, AutoModel
parsbert_config = AutoConfig.from_pretrained("HooshvareLab/bert-base-parsbert-uncased")
parsbert_tokenizer = AutoTokenizer.from_pretrained("HooshvareLab/bert-base-parsbert-uncased")
parsbert= AutoModel.from_pretrained("HooshvareLab/bert-base-parsbert-uncased")
```

```python
def preprocess_data(df):
    contexts = df['context'].tolist()
    questions = df['question'].tolist()
    start_positions = df['answer_start'].tolist()
    end_positions = df['answer_end'].tolist()

    encodings = parsbert_tokenizer(contexts, questions, truncation=True, padding=True, max_length=256, return_tensors='pt', return_offsets_mapping=True)
    start_positions_token = []
    end_positions_token = []
    
    for i in range(len(start_positions)):
        offsets = encodings['offset_mapping'][i].tolist()

        if start_positions[i] == 0 and end_positions[i] == 0:
            start_positions_token.append(0)
            end_positions_token.append(0)
        else:
            start_token_pos = [j for j, offset in enumerate(offsets) if offset[0] <= start_positions[i] < offset[1]]
            start_positions_token.append(start_token_pos[0] if start_token_pos else 255)

            end_token_pos = [j for j, offset in enumerate(offsets) if offset[0] <= end_positions[i] < offset[1]]
            end_positions_token.append(end_token_pos[0] if end_token_pos else 255)

    encodings.update({'start_positions': start_positions_token, 'end_positions': end_positions_token})

    return encodings
```

```python
import torch
train_encoding=preprocess_data(df_train)
test_encoding=preprocess_data(df_test)
val_encoding=preprocess_data(df_valid)
train_input_ids=torch.tensor(train_encoding['input_ids'])
train_attention_mask=torch.tensor(train_encoding['attention_mask'])
train_start_positions=torch.tensor(train_encoding['start_positions'])
train_end_positions=torch.tensor(train_encoding['end_positions'])
train_token_type_ids=torch.tensor(train_encoding['token_type_ids'])
test_input_ids=torch.tensor(test_encoding['input_ids'])
test_attention_mask=torch.tensor(test_encoding['attention_mask'])
test_start_positions=torch.tensor(test_encoding['start_positions'])
test_end_positions=torch.tensor(test_encoding['end_positions'])
test_token_type_ids=torch.tensor(test_encoding['token_type_ids'])
val_input_ids=torch.tensor(val_encoding['input_ids'])
val_attention_mask=torch.tensor(val_encoding['attention_mask'])
val_start_positions=torch.tensor(val_encoding['start_positions'])
val_end_positions=torch.tensor(val_encoding['end_positions'])
val_token_type_ids=torch.tensor(val_encoding['token_type_ids'])
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F
# def freeze_bert(model):
#     for name, param in model.named_parameters():
#         if 'bert' in name:
#             param.requires_grad = False
class ExtractiveQAModel(nn.Module):
    def __init__(self, bert_model,bert_config):
        super(ExtractiveQAModel, self).__init__()
        self.bert = bert_model
        self.dropout = nn.Dropout(0.1)
        self.fc = nn.Linear(bert_config.hidden_size, 2)  
    def forward(self, input_ids, attention_mask, token_type_ids):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask, token_type_ids=token_type_ids)
        sequence_output = outputs.last_hidden_state  
        sequence_output = self.dropout(sequence_output)
        logits = self.fc(sequence_output) 

        start_logits, end_logits = logits.split(1, dim=-1)
        start_logits = start_logits.squeeze(-1)
        end_logits = end_logits.squeeze(-1)

        return start_logits, end_logits
def train_epoch(model, dataloader, loss_fn, optimizer, device):
    model.train()
    total_loss = 0.0
    for batch in dataloader:
        input_ids = batch[0].to(device) 
        attention_mask = batch[1].to(device) 
        token_type_ids = batch[2].to(device) 

        start_positions = batch[3].to(device)  
        end_positions = batch[4].to(device)  

        optimizer.zero_grad()

        start_logits, end_logits = model(input_ids, attention_mask, token_type_ids)
        start_positions = start_positions.long()
        end_positions = end_positions.long()
        start_logits = start_logits.float()
        end_logits = end_logits.float()
        start_loss = loss_fn(start_logits, start_positions)
        end_loss = loss_fn(end_logits, end_positions)
        loss = start_loss + end_loss
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(dataloader)


def evaluate(model, dataloader, loss_fn, device):
    model.eval()
    total_loss = 0.0
    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch[0].to(device)  
            attention_mask = batch[1].to(device)  
            token_type_ids = batch[2].to(device)  

            start_logits, end_logits = model(input_ids, attention_mask, token_type_ids)
            start_logits = start_logits.float()
            end_logits = end_logits.float()


            start_positions = batch[3].to(device).long()  
            end_positions = batch[4].to(device).long()  

            start_loss = loss_fn(start_logits, start_positions)
            end_loss = loss_fn(end_logits, end_positions)
            loss = start_loss + end_loss

            total_loss += loss.item()

    return total_loss / len(dataloader)
def train(model, train_dataloader, val_dataloader, loss_fn, optimizer, num_epochs):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
   # model.to(device)
    
    best_val_loss = float('inf')
    for epoch in range(num_epochs):
        print(f"Epoch {epoch+1}/{num_epochs}")
        print("-------------------")
        
        # Training
        model.train()
        train_loss = train_epoch(model, train_dataloader, loss_fn, optimizer, device)
        print(f"Train Loss: {train_loss}")
        
        # Validation
        model.eval()
        val_loss = evaluate(model, val_dataloader, loss_fn, device)
        print(f"Validation Loss: {val_loss}")
        
        # Checkpoint
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), "/content/drive/MyDrive/best_model.pt")
        
        print()
class QADataset(Dataset):
    def __init__(self, input_ids, attention_mask, token_type_ids, start_positions, end_positions):
        self.input_ids = input_ids
        self.attention_mask = attention_mask
        self.token_type_ids = token_type_ids
        self.start_positions = start_positions
        self.end_positions = end_positions

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, index):
        input_id = self.input_ids[index]
        attention_mask = self.attention_mask[index]
        token_type_id = self.token_type_ids[index]
        start_position = self.start_positions[index]
        end_position = self.end_positions[index]
        return input_id, attention_mask, token_type_id, start_position, end_position
```

```python
model = ExtractiveQAModel(parsbert,parsbert_config).float()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
from torch.optim import Adam
loss_fn = nn.CrossEntropyLoss()
optimizer = Adam(model.parameters(), lr=5e-5)
train_dataset =QADataset(train_input_ids, train_attention_mask, train_token_type_ids, train_start_positions, train_end_positions)
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)

val_dataset = QADataset(val_input_ids, val_attention_mask, val_token_type_ids, val_start_positions, val_end_positions)
val_dataloader = DataLoader(val_dataset, batch_size=32, shuffle=False)

train(model, train_dataloader, val_dataloader, loss_fn, optimizer, num_epochs=1)
```

```python
from sklearn.metrics import f1_score

def calculate_f1(true_starts, true_ends, pred_starts, pred_ends):
    f1_scores = []
    for ts, te, ps, pe in zip(true_starts, true_ends, pred_starts, pred_ends):
        true_span = list(range(int(ts), int(te) + 1))
        pred_span = list(range(int(ps), int(pe) + 1))
        common = len(set(true_span) & set(pred_span))
        if common == 0:
            f1_scores.append(0)
        else:
            precision = common / len(pred_span)
            recall = common / len(true_span)
            f1_scores.append(2 * (precision * recall) / (precision + recall))
    return sum(f1_scores) / len(f1_scores)


def test(model, test_dataloader, device):
    model.eval()
    all_start_scores = []
    all_end_scores = []
    all_start_positions = []
    all_end_positions = []

    with torch.no_grad():
        for batch in test_dataloader:
            input_ids = batch[0].to(device)
            attention_mask = batch[1].to(device)
            token_type_ids = batch[2].to(device)

            start_positions = batch[3].to(device)
            end_positions = batch[4].to(device)
            all_start_positions.extend(start_positions.tolist())
            all_end_positions.extend(end_positions.tolist())

            start_scores, end_scores = model(input_ids, attention_mask, token_type_ids)
            all_start_scores.append(start_scores)
            all_end_scores.append(end_scores)

    all_start_scores = torch.cat(all_start_scores, dim=0)
    all_end_scores = torch.cat(all_end_scores, dim=0)

    # Convert model outputs to predicted positions
    predicted_start_positions = torch.argmax(all_start_scores, dim=1)
    predicted_end_positions = torch.argmax(all_end_scores, dim=1)

    # Calculate metrics
    exact_matches = ((predicted_start_positions.cpu() == torch.tensor(all_start_positions)) & 
                     (predicted_end_positions.cpu() == torch.tensor(all_end_positions))).float().mean()

    f1 = calculate_f1(all_start_positions, all_end_positions, 
                      predicted_start_positions.cpu().tolist(), predicted_end_positions.cpu().tolist())

    print(f"Exact Match (EM): {exact_matches.item()}")
    print(f"F1 Score: {f1}")

    return all_start_scores, all_end_scores


model = ExtractiveQAModel(parsbert,parsbert_config)
model.load_state_dict(torch.load('/content/drive/MyDrive/best_model.pt'))
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

test_dataset = QADataset(test_input_ids, test_attention_mask, test_token_type_ids, test_start_positions, test_end_positions)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False)

start_scores, end_scores = test(model, test_dataloader, device)
```

```python
from transformers import AutoConfig, AutoTokenizer, AutoModel

albert_config = AutoConfig.from_pretrained('HooshvareLab/albert-fa-zwnj-base-v2')
albert_tokenizer = AutoTokenizer.from_pretrained('HooshvareLab/albert-fa-zwnj-base-v2')
albert=AutoModel.from_pretrained('HooshvareLab/albert-fa-zwnj-base-v2')
```

```python
def preprocess_data(df):
    contexts = df['context'].tolist()
    questions = df['question'].tolist()
    start_positions = df['answer_start'].tolist()
    end_positions = df['answer_end'].tolist()

    encodings = albert_tokenizer(contexts, questions, truncation=True, padding=True, max_length=256, return_tensors='pt', return_offsets_mapping=True)
    start_positions_token = []
    end_positions_token = []
    
    for i in range(len(start_positions)):
        offsets = encodings['offset_mapping'][i].tolist()

        if start_positions[i] == 0 and end_positions[i] == 0:
            start_positions_token.append(0)
            end_positions_token.append(0)
        else:
            start_token_pos = [j for j, offset in enumerate(offsets) if offset[0] <= start_positions[i] < offset[1]]
            start_positions_token.append(start_token_pos[0] if start_token_pos else 255)

            end_token_pos = [j for j, offset in enumerate(offsets) if offset[0] <= end_positions[i] < offset[1]]
            end_positions_token.append(end_token_pos[0] if end_token_pos else 255)

    encodings.update({'start_positions': start_positions_token, 'end_positions': end_positions_token})

    return encodings
```

```python
import torch
train_encoding=preprocess_data(df_train)
test_encoding=preprocess_data(df_test)
val_encoding=preprocess_data(df_valid)
train_input_ids=torch.tensor(train_encoding['input_ids'])
train_attention_mask=torch.tensor(train_encoding['attention_mask'])
train_start_positions=torch.tensor(train_encoding['start_positions'])
train_end_positions=torch.tensor(train_encoding['end_positions'])
train_token_type_ids=torch.tensor(train_encoding['token_type_ids'])
test_input_ids=torch.tensor(test_encoding['input_ids'])
test_attention_mask=torch.tensor(test_encoding['attention_mask'])
test_start_positions=torch.tensor(test_encoding['start_positions'])
test_end_positions=torch.tensor(test_encoding['end_positions'])
test_token_type_ids=torch.tensor(test_encoding['token_type_ids'])
val_input_ids=torch.tensor(val_encoding['input_ids'])
val_attention_mask=torch.tensor(val_encoding['attention_mask'])
val_start_positions=torch.tensor(val_encoding['start_positions'])
val_end_positions=torch.tensor(val_encoding['end_positions'])
val_token_type_ids=torch.tensor(val_encoding['token_type_ids'])
```

```python
def albert_train(model, train_dataloader, val_dataloader, loss_fn, optimizer, num_epochs):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
   # model.to(device)
    
    best_val_loss = float('inf')
    for epoch in range(num_epochs):
        print(f"Epoch {epoch+1}/{num_epochs}")
        print("-------------------")
        
        # Training
        model.train()
        train_loss = train_epoch(model, train_dataloader, loss_fn, optimizer, device)
        print(f"Train Loss: {train_loss}")
        
        # Validation
        model.eval()
        val_loss = evaluate(model, val_dataloader, loss_fn, device)
        print(f"Validation Loss: {val_loss}")
        
        # Checkpoint
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), "/content/drive/MyDrive/albert_model.pt")
        
        print()
```
